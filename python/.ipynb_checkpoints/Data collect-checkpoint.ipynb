## Looking for relevant datasets

Estimated percentage of outpatient **doctor visits primarily about COVID-related symptoms**, based on data from health system partners, smoothed in time using a Gaussian linear smoother with systematic day-of-week effects removed

```python
ca_dv_smoothed_adj_cli = covidcast.signal("doctor-visits", "smoothed_adj_cli",
                                          start_date, end_date,
                                          "state",geo_values=["CA"])
ca_dv_smoothed_adj_cli.reset_index(inplace=True)
del ca_dv_smoothed_adj_cli['index']
```

Number of new confirmed COVID-19 cases, daily

```python
ca_ic_confirmed_7dav_incidence_num = covidcast.signal("indicator-combination", "confirmed_7dav_incidence_num",
                        start_date, end_date,
                        "state", ["CA"])
ca_ic_confirmed_7dav_incidence_num.reset_index(inplace=True)
del ca_ic_confirmed_7dav_incidence_num['index']
```

Every day, Delphi surveys tens of thousands of Facebook users, asking a broad set of COVID-related questions, including whether they, or anyone in their household, are **currently experiencing COVID-related symptoms**.

```python
ca_fb_smoothed_cli = covidcast.signal("fb-survey", "smoothed_cli",
                        start_date, end_date,
                        "state", ["CA"])
ca_fb_smoothed_cli.reset_index(inplace=True)
del ca_fb_smoothed_cli['index']
```

 We also ask them if they know anyone in their **local community who has COVID-like** or flu-like **symptoms**

```python

ca_fb_smoothed_cmnty_cli = covidcast.signal("fb-survey", "smoothed_hh_cmnty_cli",
                        start_date, end_date,
                        "state", ["CA"])
ca_fb_smoothed_cmnty_cli.reset_index(inplace=True)
del ca_fb_smoothed_cmnty_cli['index']
```

**Google Symptoms Searches**, Delphi obtains the sum of Google search volumes for anosmia (loss of smell) related searches and ageusia (loss of taste) related searches in each area

```python
ca_google_symptom_smoothed_search = covidcast.signal("google-symptoms", "sum_anosmia_ageusia_smoothed_search",
                        start_date, end_date,
                        "state", ["CA"])
ca_google_symptom_smoothed_search.reset_index(inplace=True)
del ca_google_symptom_smoothed_search['index']
```

We estimate the percentage of **new hospital admissions each day** that are related to COVID-19

```python
ca_ha_smoothed_adj_covid19_from_claims = covidcast.signal("hospital-admissions", "smoothed_adj_covid19_from_claims",
                        start_date, end_date,
                        "state", ["CA"])
ca_ha_smoothed_adj_covid19_from_claims.reset_index(inplace=True)
del ca_ha_smoothed_adj_covid19_from_claims['index']
```

### Test some signals

```python
def test_signal(data_source, signal):
    data = covidcast.signal(data_source,signal,
                        start_date, end_date,
                        "state", ["CA"])
    data.reset_index(inplace=True)
    del data['index']
    
    return data.value, signal
```

```python
values, signal = test_signal("fb-survey", "smoothed_cli")

plt.plot(values)
plt.title(signal)
plt.show()
```

```python
values, signal = test_signal("fb-survey", "smoothed_hh_cmnty_cli")

plt.plot(values)
plt.title(signal)
plt.show()
```

## Merge relevant datasets

 - ca_dv_smoothed_adj_cli

```python
relevant_datasets = [ca_dv_smoothed_adj_cli, 
                     ca_ic_confirmed_7dav_incidence_num, 
                     ca_fb_smoothed_cli, 
                     ca_fb_smoothed_cmnty_cli,
                     ca_google_symptom_smoothed_search,
                     ca_ha_smoothed_adj_covid19_from_claims]
```

```python
merged_relevant_dfs = pd.DataFrame()
```

```python
timediff_in_days = (end_date - start_date).days + 1
timediff_in_days
```

```python
merged_relevant_dfs['time_value'] = pd.date_range(start_date, periods=timediff_in_days, freq="d")
```

```python
for df in relevant_datasets:
    selected_cols = df[['time_value', 'value']]
    col_name = df.loc[0].data_source +'_'+df.loc[0].signal
    
    selected_cols.columns = ['time_value', col_name]

    merged_relevant_dfs = pd.merge(merged_relevant_dfs,selected_cols,on='time_value',how='left')
```

```python
selected = normalized_df.reset_index()[['doctor-visits_smoothed_adj_cli']]
```

```python
def custom_fill_na(selected):
    for index, num in selected.iterrows():
        if np.isnan(num[0]) and index > 4:

            nans = 0
            for i in range(index, selected.shape[0]):
                if np.isnan(selected.loc[i][0]):
                    nans += 1
                else:
                    break

            last_known_ind = index - 1
            first_known_ind = index + nans
            sdiff = 3

            diff_m_43 = selected.loc[last_known_ind - 3][0] - selected.loc[last_known_ind - 2][0]
            diff_m_32 = selected.loc[last_known_ind - 2][0] - selected.loc[last_known_ind - 1][0]
            diff_m_21 = selected.loc[last_known_ind - 1][0] - selected.loc[last_known_ind][0]

            diff_m_12 = selected.loc[first_known_ind][0]    - selected.loc[first_known_ind + 1][0]
            diff_m_23 = selected.loc[first_known_ind + 1][0]- selected.loc[first_known_ind + 2][0]
            diff_m_34 = selected.loc[first_known_ind + 2][0]- selected.loc[first_known_ind + 3][0]

            indexes = np.arange(index - sdiff,index+nans+sdiff)

            diffs = np.concatenate((
                [diff_m_43, diff_m_32, diff_m_21],
                np.zeros(nans),
                [diff_m_12, diff_m_23, diff_m_34]             
                                   ))

            lindiff = (selected.loc[first_known_ind][0] - selected.loc[last_known_ind][0]) / nans 
            itnum = 0
            for i in range(index, index + nans):
                itnum += 1
                replace_value = selected.loc[last_known_ind][0] + itnum * lindiff + sum( np.exp( -(indexes - i)**2 / (2 * **2) ) * diffs )
                print( replace_value )
                selected.loc[i][0] = replace_value
```
